```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"E:\Machine-learning\Amoozeshgah-Tahlil-Dadeh\s4\Exercise\diabetes.csv")
df = pd.DataFrame(data)
df
```

```python
columns = df.columns.str.lower()
df.columns = columns
df
```

```python
df.describe()
```

```python
diabates = df[df["outcome"] == 1]
diabates.count()
```

```python
non_diabates = df[df["outcome"] == 0]
non_diabates.count()
```

```python
x = np.arange(20,81,1)
mean_diabates = diabates["glucose"].mean()
mean_nondiabates = non_diabates["glucose"].mean()
mean_diabates =pd.Series([mean_diabates]*len(x))
mean_nondiabates = pd.Series([mean_nondiabates]*len(x))
```

```python
plt.figure(figsize=(8,6), dpi=80)
plt.scatter(non_diabates["age"], non_diabates["glucose"], c="red", label="non_diabates")
plt.scatter(diabates["age"], diabates["glucose"], c="blue", label="diabates")
plt.plot(x, mean_nondiabates, linewidth=2, color="red", label="mean_nondiabates" )
plt.plot(x, mean_diabates, linewidth=2, color="blue", label="mean_diabates" )
plt.xlim(20,80)
plt.xlabel("Age", fontsize=10)
plt.ylabel("Glucose", fontsize=10)
plt.xticks(fontsize=10)
plt.legend(fontsize=10)

plt.show()
```

```python
#The chart above depicts the levels of glucose in diabetic and non-diabetic patients according to age groups.
#Considering the figure, the average level of glucose in diabetic patients is higher than that in non-diabetic patients.
#The data points with a glucose level of 0 can be considered as suspicious or erroneous samples and can be excluded from further analyses.
```

```python
fig, axs= plt.subplots(2,1, figsize=(8,8) )
axs[0].hist(diabates["glucose"], bins=20, color="blue", label="diabates")
axs[0].set_xlabel("glucose", fontsize=12)
axs[0].set_ylabel("Frequency", fontsize=12)
axs[0].set_ylim(0,40)
#axs[0].set_title("Diabates")
axs[0].legend(fontsize=12)

axs[1].hist(non_diabates["glucose"], bins=20, color="red", label="non diabates")
axs[1].set_xlabel("glucose", fontsize=12)
axs[1].set_ylabel("Frequency", fontsize=12)
axs[1].set_ylim(0,100)
#axs[1].set_title("Non Diabates")
axs[1].legend(fontsize=12)
fig.tight_layout()
#fig.legend()
#fig.set_figheight(6)
#fig.set_figwidth(6)

plt.show()
```

```python
#The histogram above illustrates the frequency distribution of glucose levels in diabetic and non-diabetic patients. 
#While non-diabetic patients predominantly exhibit glucose levels between 90 to 120, diabetic patients show a majority distribution ranging from 100 to 190. 
#It's evident that glucose levels are higher in diabetic patients.
```
